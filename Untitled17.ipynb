```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def apply_filter(image, filter_type):
    if filter_type == 'grayscale':
        return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    elif filter_type == 'blur':
        return cv2.GaussianBlur(image, (11, 11), 0)
    elif filter_type == 'edges':
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        return cv2.Canny(gray, 100, 200)
    else:
        return image

def adjust_brightness_contrast(image, brightness=0, contrast=0):
    alpha = (contrast + 100) / 100.0
    beta = brightness
    return cv2.addWeighted(image, alpha, np.zeros(image.shape, image.dtype), 0, beta)

# Upload image file
uploaded = files.upload()

# Load the image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# Ask for filter type
filter_type = input("Enter filter type (grayscale, blur, edges): ")

# Apply a filter
filtered_image = apply_filter(image, filter_type)

# Ask for brightness and contrast values
brightness = int(input("Enter brightness value (-100 to 100): "))
contrast = int(input("Enter contrast value (-100 to 100): "))

# Adjust brightness and contrast
adjusted_image = adjust_brightness_contrast(filtered_image, brightness=brightness, contrast=contrast)

# Display the original and manipulated images
cv2_imshow(image)
cv2_imshow(adjusted_image)

# Save the original and manipulated images
save_original = input("Save original image? (y/n): ")
if save_original.lower() == 'y':
    cv2.imwrite("/content/original_image.jpg", image)
    print("Original image saved.")

save_adjusted = input("Save adjusted image? (y/n): ")
if save_adjusted.lower() == 'y':
    cv2.imwrite("/content/adjusted_image.jpg", adjusted_image)
    print("Adjusted image saved.")

# Display histograms
plt.figure(figsize=(15, 6))

plt.subplot(2, 2, 1)
plt.title("Original Image Histogram")
plt.hist(image.ravel(), 256, [0, 256])

plt.subplot(2, 2, 2)
plt.title("Adjusted Image Histogram")
plt.hist(adjusted_image.ravel(), 256, [0, 256])

plt.show()

# Save histograms as an image
histogram_img_path = "/content/histogram.jpg"
plt.savefig(histogram_img_path)
print("Histogram image saved.")

# Download saved files
file_paths = ["/content/original_image.jpg", "/content/adjusted_image.jpg", histogram_img_path]
for file_path in file_paths:
    files.download(file_path)
```
